<div id="boolean">

<h3 style="text-align:center; color:skyblue; text-decoration: underline  ; font-weight: bold">Boolean</h3>

<p>It is a data type that has two values: <code>True (or 1)</code> and <code>False (or 0)</code>. </p>

<p>Note that <code>True</code> and <code>False</code> are case-sensitive</p>

</div>


<div id="complex">

<h3 style="text-align:center; color:skyblue; text-decoration: underline  ; font-weight: bold">Complex</h3>

<p>It is a data type that has two parts: <code>real</code> and <code>imaginary (represented by j)</code>. </p>

<p>Which are separated by the <code>+</code> or <code>-</code> sign</p>
<p>For example: <code>3+4j</code> or <code>3-4j</code> </p>
