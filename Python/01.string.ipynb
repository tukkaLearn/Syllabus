<div id="slicing">

<h3 style="text-align:center; color:skyblue ; text-decoration: underline ;font-weight: bold">Slicing</h3>  
<p>It is a way to access multiple characters in a string. 
<span style="color:yellow"> Syntax: string[start:end:step] </span>
</p>
<i>Default value of start is 0 </i>

<i>Default value of end is length of string or said -1</i>

<i>Default value of step is 1</i>

<h3 style="text-align:center; color:skyblue ; text-decoration: underline ;font-weight: bold">String Methods</h3> 
</div>

<div>
<ol style="color:#0fb3ff ; display:flex ; flex-direction: column ; gap:15px">

<li>
<abbr title="Returns length of string">len() note: it is a built-in function not a method </abbr>
</li>

<li>
<abbr title="Returns a copy of the string with leading and trailing whitespace removed">strip</abbr>
</li>

<li>
<abbr title="Returns a copy of the string with leading whitespace removed">lstrip</abbr>
</li>

<li>
<abbr title="Returns a copy of the string with trailing whitespace removed">rstrip</abbr>
</li>

<li>
<abbr title="Returns a copy of the string with uppercase characters converted to lowercase">lower</abbr>
</li>

<li>
<abbr title="Returns a copy of the string with lowercase characters converted to uppercase">upper</abbr>
</li>

<li>
<abbr title="Returns a copy of the string with the first character converted to uppercase and the rest to lowercase">capitalize</abbr>
</li>

<li>
<abbr title="Returns a copy of the string with the first character converted to uppercase">title</abbr>
</li>

<li>
<abbr title="Returns a copy of the string with uppercase characters converted to lowercase and lowercase characters converted to uppercase">swapcase</abbr>
</li>

<li><abbr title="Splits the string into a list using a separator">split</abbr></li>

  <li><abbr title="Checks if all characters are lowercase">islower</abbr></li>
  <li><abbr title="Checks if all characters are uppercase">isupper</abbr></li>
  <li><abbr title="Checks if all characters in the string are alphabetic">isalpha</abbr></li>
  <li><abbr title="Checks if all characters in the string are digits">isdigit</abbr></li>
  <li><abbr title="Checks if all characters in the string are alphanumeric">isalnum</abbr></li>
 <li><abbr title="Checks if the string starts with the specified prefix">startswith</abbr></li>
  <li><abbr title="Checks if the string ends with the specified suffix">endswith</abbr></li>
  <li><abbr title="Replaces a specified value with another">replace</abbr></li>
  <li><abbr title="Returns the lowest index where the substring is found">find</abbr></li>
  <li><abbr title="Returns the lowest index where the substring is found from right side">rfind</abbr></li>
  <li><abbr title="Counts the number of times a substring occurs">count</abbr></li>
  <li><abbr title="Fills the string with leading zeros to reach the specified length">zfill</abbr></li>
  <li><abbr title="Centers the string in a field of given width">center</abbr></li>
  <li><abbr title="Formats the string using placeholders">format</abbr></li>
  <li><abbr title="Splits the string into a 3-part tuple: before, separator, after">partition</abbr></li>
  <li><abbr title="Same as partition() but splits from the right">rpartition</abbr></li>
  <li><abbr title="Removes a specific prefix from the string (Python 3.9+)">removeprefix</abbr></li>
  <li><abbr title="Removes a specific suffix from the string (Python 3.9+)">removesuffix</abbr></li>
  <li><abbr title="Joins elements of a list into a string using a separator">join</abbr></li>
  <li><abbr title="give ASCII value od Given Character">ord</abbr></li>
  <li><abbr title="give character of corresponding ASCII value ">chr</abbr></li>
