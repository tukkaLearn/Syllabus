<h3 style="text-align:center; color:skyblue; text-decoration: underline  ; font-weight: bold">Hypothesis</h3>


```python
# Z-test (when σ is known + large sample n ≥ 30)
# Situation
# μFactory claims new bulb lasts  = 1000 hours.
# Population std dev σ = 50 hours (known from past).
# We tested 40 bulbs → sample mean = 985 hours.
```

```python
import numpy as np
from scipy import stats

H0 = 42000
mu   = 42500 #H1
sigma = 4000
N     = 81

z = (H0 - mu) / (sigma / np.sqrt(N))

p_value = 2 * (1 - stats.norm.cdf(abs(z)))   

print(f"Z = {z:.2f}")
print(f"P-value = {p_value:.4f}")
if p_value>0.05:
    print("Fail To Reject")
else:
    print("Not Accepted")
# p_value>0.05 H0 fail to reject
```

```python
# One-sample t-test (σ unknown + small sample)
# Same claim (1000 hours), but now we only tested 9 bulbs
# Data: 980, 1010, 995, 1025, 990, 1005, 970, 1030, 1015
```

```python
# Chi-square test (for independence / goodness of fit)
# Situation
# Survey: Do boys and girls prefer same subjects?
```

```python
# One-way ANOVA (compare 3+ groups)
# Situation
# Do three teaching methods give same exam scores?

# Method A (8 students): 72, 75, 78, 80, 82, 85, 88, 90
# Method B (8 students): 65, 68, 70, 72, 75, 78, 80, 82
# Method C (8 students): 80, 82, 85, 88, 90, 92, 95, 98
```

```python
from scipy import stats
import numpy as np

data = np.array([980,1010,995,1025,990,1005,970,1030,1015])

# 90% confidence
print(stats.t.interval(0.90, df=len(data)-1, loc=np.mean(data), scale=stats.sem(data)))
# → wider interval

# 95% confidence (default everyone uses)
print(stats.t.interval(0.95, df=len(data)-1, loc=np.mean(data), scale=stats.sem(data)))
# → normal interval

# 99% confidence
print(stats.t.interval(0.99, df=len(data)-1, loc=np.mean(data), scale=stats.sem(data)))
# → very wide interval (super safe)
```
