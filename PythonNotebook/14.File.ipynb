<div id="file">

<h3 style="text-align:center; color:skyblue; text-decoration: underline  ; font-weight: bold">File Handling</h3>

<table border="1" style="border-collapse: collapse; text-align: left; width: 100%;">
  <thead style="background-color:rgb(0, 0, 0);">
    <tr>
      <th>Mode</th>
      <th>Description</th>
      <th>Use Case</th>
    </tr>
  </thead>
  <tbody>
    <tr>
      <td><code>'r'</code></td>
      <td>Read mode (default). Opens file for reading, error if file does not exist.</td>
      <td>Reading existing files</td>
    </tr>
    <tr>
      <td><code>'w'</code></td>
      <td>Write mode. Creates new file or truncates existing one.</td>
      <td>Overwriting or creating new files</td>
    </tr>
    <tr>
      <td><code>'a'</code></td>
      <td>Append mode. Writes at end of file if exists, creates if not.</td>
      <td>Adding logs or data without overwriting</td>
    </tr>
    <tr>
      <td><code>'r+'</code></td>
      <td>Read and write mode. File must exist.</td>
      <td>Update existing files</td>
    </tr>
    <tr>
      <td><code>'w+'</code></td>
      <td>Write and read mode. Overwrites file.</td>
      <td>Clear file and update</td>
    </tr>
    <tr>
      <td><code>'a+'</code></td>
      <td>Append and read mode.</td>
      <td>Read and add without deleting content</td>
    </tr>
    <tr>
      <td><code>'b'</code></td>
      <td>Binary mode. Read/write in binary format.</td>
      <td>For files like images, audio, etc.</td>
    </tr>
  </tbody>
</table>
