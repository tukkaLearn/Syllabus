    <h3>List Methods And Useful Function</h3> 
</div>

<div>
    <ol>
        <li>
            <abbr title="Adds an item to the end of the list">append()</abbr> 
        </li>
        <li>
            <abbr title="Inserts an item at a specific index in the list">insert()</abbr> 
        </li>
        <li>
            <abbr title="Adds all elements of an iterable to the list">extend()</abbr>
        </li>
        <li>
            <abbr title="Removes and returns an item at a given index, defaulting to the last item">pop()</abbr> 
        </li>
        <li>
            <abbr title="Removes the first occurrence of a specified value from the list">remove()</abbr> 
        </li>
        <li>
            <abbr title="Returns the index of the first occurrence of a value in the list">index()</abbr>
        </li>
        <li>
            <abbr title="Counts the number of occurrences of a value in the list">count()</abbr>
        </li>
        <li>
            <abbr title="Sorts the list in ascending order">sort()</abbr>
        </li>
        <li>
            <abbr title="Reverses the order of the list">reverse()</abbr>
        </li>
        <li>
            <abbr title="Creates a shallow copy of the list">copy()</abbr>

         <li>
            <abbr title="Removes all elements from the list">clear()</abbr>
