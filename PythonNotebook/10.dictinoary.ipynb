<h3 style="text-align:center; color:skyblue; text-decoration: underline  ; font-weight: bold">Dictionary</h3>

<p>A dictionary consists of keys and values. It is helpful to compare a dictionary to a list. Instead of being indexed numerically like a list, dictionaries have keys. These keys are the keys that are used to access values within a dictionary.  
</p>
<br/>

<ol style="display: flex; flex-direction: column; gap: 10px;font-size: 18px">
   <li>Unordered (till Python 3.6)</li>
   <li>Ordered (from Python 3.7 onwards)</li>
   <li>Mutable</li>
   <li>Indexed using keys (not positions)</li>
   <li>Duplicated keys are not allowed</li>

</ol>

<div id="methods">
    <h3>Dictionary Methods and Operations</h3>
</div>

<div>
    <ol>
      <li><abbr title="Returns a list of keys">dict.keys()</abbr></li> 
      <li><abbr title="Returns a list of values">dict.values()</abbr></li> 
      <li><abbr title="Returns a list of key-value tuples">dict.items()</abbr></li> 
      <li><abbr title="Returns value if key exists, else default">dict.get(key, default)</abbr></li> <li><abbr title="Add key with default if missing">dict.setdefault(key, default)</abbr></li> 
      <li><abbr title="Update dict with another dict or pairs">dict.update(other_dict)</abbr></li>
       <li><abbr title="Removes and returns value by key">dict.pop(key)</abbr></li> 
       <li><abbr title="Removes key">del </abbr></li> 
       <li><abbr title="Removes the last inserted key-value (Python 3.7+)">dict.popitem()</abbr></li> 
       <li><abbr title="Remove all elements">dict.clear()</abbr></li>
    </ol>
