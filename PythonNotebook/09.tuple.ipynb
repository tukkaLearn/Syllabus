<h3 style="text-align:center; color:skyblue; text-decoration: underline  ; font-weight: bold">tuple</h3>
Ordered collection of items, similar to a list, but with a key difference: **tuples are immutable**. Once a tuple is created, its content cannot be changed.

### Key Properties of Tuples

1. **Ordered** : Items in a tuple have a defined order.

2. **Immutable** : Tuples cannot be modified after creation.

3. **Allow Duplicate Values** : Tuples can store repeated values.

4. **Heterogeneous** : Tuples can store different types of data.

5. **Indexed** : Elements can be accessed using indices, starting from 0.
