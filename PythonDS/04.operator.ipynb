<div id="arithmetic">

<h3 style="text-align:center; color:skyblue; text-decoration: underline  ; font-weight: bold">Arithmetic</h3>

<p>Arithmetic operators are used to perform common mathematical operations such as addition, subtraction, multiplication, and division.</p>

<h5>Types ofArithmetic Operators</h4>
<ul style="display: flex; flex-direction: column; gap: 10px;font-size: 18px">
    <li> <code>Addition <b>+</b></code></li>
    <li> <code>Subtraction <b>-</b></code></li>
    <li> <code>Multiplication <b>*</b></code></li>
    <li> <code>Division <b>/</b></code></li>
    <li> <code>Floor Division <b>//</b></code></li>
    <li> <code>Modulo/Remainder <b>%</b></code></li>
    <li> <code>Exponent <b>**</b></code></li>
</ul>

<p>The Associativity of Arithmetic Operators is from <i>left to right.</i> </p>

</div>

<p>Assignment operators are used to assign values to variables. Note the associativity of the assignment operator is from <i>right to left.</i></p>

<ul style="display: flex; flex-direction: column; gap: 10px;font-size: 18px">
    <li> <code>Assignment<b> =</b></code></li>
</ul>

</div>

<p>Relational operators are used to compare two values. <u>Output is either True or False.</u> </p>

<div id="logical">

<h3 style="text-align:center; color:skyblue; text-decoration: underline  ; font-weight: bold">Logical</h3>

<p>Logical operators are used to combine conditional statements.</p>

<ul style="display: flex; flex-direction: column; gap: 10px;font-size: 18px">
    <li> <code> and </code> <u>Logical and , if both conditions are true then result is true </u> </li>
    <li> <code> or </code> <u>Logical or , if at least one condition is true then result is true </u> </li>
    <li> <code> not </code> <u>Logical not , if condition is false then result is true </u> </li>

</ul>

<mark>It not Always Return True or False it returns the actual value of the operand where the decision was made. </mark>
<br/>

</div>

<p> Associativity of Logical Operators is from <i>left to right.</i> </p>

<h4>Truthy and Falsy</h4>

| Value            | Truthy/Falsy |
| ---------------- | ------------ |
| `0`              | Falsy        |
| `''`             | Falsy        |
| `[]`             | Falsy        |
| `None`           | Falsy        |
| `False`          | Falsy        |
| Any number ≠ 0   | Truthy       |
| Non-empty string | Truthy       |
| Non-empty list   | Truthy       |

</div>


<h4 style="text-align:center"> and with Truthy/Falsy <h4>

<div style="display:flex ; justify-content: center">

| Expression        | Result | Explanation                         |
| ----------------- | ------ | ----------------------------------- |
| `True and True`   | `True` | Both True → returns last (`True`)   |
| `1 and 2`         | `2`    | Both truthy → returns last (`2`)    |
| `2 and 3 and 4`   | `4`    | All truthy → returns last           |
| `0 and 5`         | `0`    | `0` is falsy → returned immediately |
| `'' and 'Python'` | `''`   | Empty string is falsy               |
| `5 and 0 and 1`   | `0`    | Stops at `0` (first falsy)          |

</div>


<div id="bitwise">

<h3 style="text-align:center; color:skyblue; text-decoration: underline  ; font-weight: bold">Bitwise</h3>

<p>Bitwise operators are used to perform bitwise operations on integers.</p>

<ul style="display: flex; flex-direction: column; gap: 10px;font-size: 18px">
    <li> <code> & </code> <u>Bitwise AND , if both bits are 1 then result is 1 </u></li>
    <li> <code> | </code> <u>Bitwise OR , if either bit is 1 then result is 1 </u></li>
    <li> <code> ^ </code> <u>Bitwise XOR , if both bits are same then result is 0 , if both bits are different then result is 1</u></li>
    <li> <code> ~ </code> <u>Bitwise NOT , if bit is 1 then result is 0 , if bit is 0 then result is 1</u></li>
    <li> <code> << </code> <u>Bitwise left shift </u></li>
    <li> <code> >> </code> <u>Bitwise right shift </u></li>

</ul>
<br/>

<img src="https://media.geeksforgeeks.org/wp-content/uploads/20220922145839/BItwiseoperatortruthtable-300x197.png" width="500">

</div>
<p>Associativity of Bitwise Operators is from <i>right to left.</i> </p>
</div>
