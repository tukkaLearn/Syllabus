<h3 style="text-align:center; color:skyblue; text-decoration: underline  ; font-weight: bold">Set Methods And Operations </h3>

<ol style="color:#0fb3ff ; display:flex ; flex-direction: column ; gap:15px">

<li><abbr title="Adds a single item to the set">add()</abbr></li>
<li><abbr title="Adds multiple items to the set">update()</abbr></li>
<li><abbr title="Removes an item (raises error if not present)">remove()</abbr></li>
<li><abbr title="Removes an item (no error if not present)">discard()</abbr></li>
<li><abbr title="Removes a random item from the set">pop()</abbr></li>
<li><abbr title="Clears all items in the set">clear()</abbr></li>
<li><abbr title="Returns a shallow copy of the set">copy()</abbr></li>

<li><abbr title="Returns the common items in both sets">intersection() , &</abbr></li>
<li><abbr title="Returns the union of two sets">union() , |</abbr></li>
<li><abbr title="Returns items in first set but not in second">difference() , -</abbr></li>
<li><abbr title="Returns items not common in both sets">symmetric_difference() , ^</abbr></li>
<li><abbr title="Returns True if the first set is a superset of the second">issubset()</abbr></li>
<li><abbr title="Returns True if the first set is a proper superset of the second">issuperset()</abbr></li>
<li><abbr title="Returns True if the two sets have no items in common">isdisjoint()</abbr></li>

</ol>

</div>

<div id="issubset">

<h3 style="text-align:center; color:skyblue; text-decoration: underline  ; font-weight: bold">IsSubset,IsSuperset, IsDisjoint</h3>

<code>issubset</code>

<p>checks if all elements of one set are present in another set. It returns True</p>

<code>issuperset</code>

<p>checks if one set contains all elements of another set. It returns True</p>

<br/>

<code>isdisjoint</code>

<p> This method checks if two sets have no common elements. It returns True</p>
