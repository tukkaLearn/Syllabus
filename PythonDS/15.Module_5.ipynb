<div id="re">

<h3 style="text-align:center; color:skyblue; text-decoration: underline  ; font-weight: bold">re</h3>

<div>
<table border="1" cellpadding="8" cellspacing="0" style="border-collapse: collapse; width: 100%; font-family: Arial, sans-serif;">
  <thead style="background-color:rgb(24, 22, 22);">
    <tr>
      <th style="text-align: left;">Symbol</th>
      <th style="text-align: left;">Description</th>
      <th style="text-align: left;">Example</th>
      <th style="text-align: left;">Matches</th>
    </tr>
  </thead>
  <tbody>
    <tr>
      <td><code>[abc]</code></td>
      <td>Character class: matches any one character listed</td>
      <td><code>[aeiou]</code></td>
      <td>any vowel</td>
    </tr>
    <tr>
      <td><code>[^abc]</code></td>
      <td>Negated character class: matches any character <strong>not</strong> listed</td>
      <td><code>[^0-9]</code></td>
      <td>any non-digit character</td>
    </tr>
    <tr>
      <td><code>\\</code></td>
      <td>Escape character: matches a literal backslash or special char</td>
      <td><code>\\.</code></td>
      <td>a literal dot <code>.</code></td>
    </tr>
    <tr>
      <td><code>.</code></td>
      <td>Matches any single character except newline</td>
      <td><code>a.b</code></td>
      <td><code>aab</code>, <code>acb</code>, <code>a9b</code></td>
    </tr>
    <tr>
      <td><code>*</code></td>
      <td>Matches zero or more repetitions of the previous character</td>
      <td><code>ca*t</code></td>
      <td><code>ct</code>, <code>cat</code>, <code>caaaat</code></td>
    </tr>
    <tr>
      <td><code>+</code></td>
      <td>Matches one or more repetitions of the previous character</td>
      <td><code>go+gle</code></td>
      <td><code>gogle</code>, <code>google</code>, <code>gooogle</code></td>
    </tr>
    <tr>
      <td><code>?</code></td>
      <td>Matches zero or one repetition (makes preceding optional)</td>
      <td><code>colou?r</code></td>
      <td><code>color</code>, <code>colour</code></td>
    </tr>
    <tr>
      <td><code>^</code></td>
      <td>Matches the start of the string or line</td>
      <td><code>^Hello</code></td>
      <td><code>Hello there</code> (start match)</td>
    </tr>
    <tr>
      <td><code>$</code></td>
      <td>Matches the end of the string or line</td>
      <td><code>end$</code></td>
      <td><code>The end</code> (end match)</td>
    </tr>
    <tr>
      <td><code>{m}</code></td>
      <td>Matches exactly <code>m</code> repetitions of the preceding element</td>
      <td><code>a{3}</code></td>
      <td><code>aaa</code></td>
    </tr>
    <tr>
      <td><code>{m,n}</code></td>
      <td>Matches between <code>m</code> and <code>n</code> repetitions</td>
      <td><code>a{2,4}</code></td>
      <td><code>aa</code>, <code>aaa</code>, <code>aaaa</code></td>
    </tr>
    <tr>
      <td><code>( ... )</code></td>
      <td>Groups patterns or captures matched substring</td>
      <td><code>(ab)+</code></td>
      <td><code>ab</code>, <code>abab</code>, <code>ababab</code></td>
    </tr>
    <tr>
      <td><code>|</code></td>
      <td>OR operator: matches either pattern on left or right</td>
      <td><code>cat|dog</code></td>
      <td><code>cat</code> or <code>dog</code></td>
    </tr>
    <tr>
      <td><code>\d</code></td>
      <td>Matches any digit (0-9)</td>
      <td><code>\d{3}</code></td>
      <td><code>123</code>, <code>456</code>, <code>007</code></td>
    </tr>
    <tr>
      <td><code>\D</code></td>
      <td>Matches any non-digit character</td>
      <td><code>\D+</code></td>
      <td><code>abc</code>, <code>!@#</code></td>
    </tr>
    <tr>
      <td><code>\w</code></td>
      <td>Matches any alphanumeric character or underscore</td>
      <td><code>\w+</code></td>
      <td><code>hello123</code>, <code>_var</code></td>
    </tr>
    <tr>
      <td><code>\W</code></td>
      <td>Matches any non-word character</td>
      <td><code>\W+</code></td>
      <td><code>!@#</code>, spaces</td>
    </tr>
    <tr>
      <td><code>\s</code></td>
      <td>Matches any whitespace character (space, tab, newline)</td>
      <td><code>\s+</code></td>
      <td>space, tab, newline</td>
    </tr>
    <tr>
      <td><code>\S</code></td>
      <td>Matches any non-whitespace character</td>
      <td><code>\S+</code></td>
      <td><code>hello</code>, <code>123</code></td>
    </tr>
    <tr>
      <td><code>\b</code></td>
      <td>Matches a word boundary</td>
      <td><code>\bcat\b</code></td>
      <td><code>cat but not cats or cat.</code></td>
    </tr>
  </tbody>
</table>

</div>

<li><abbr title="Matches pattern only at the beginning of the string">re.match(pattern, string)</abbr></li>
<li><abbr title="Finds all non-overlapping matches of pattern in string">re.findall(pattern, string)</abbr></li>
<li><abbr title="Returns an iterator yielding match objects over all matches">re.finditer(pattern, string)</abbr></li>

<li><abbr title="Flags to modify matching behavior: IGNORECASE, MULTILINE, DOTALL">re.IGNORECASE, re.MULTILINE, re.DOTALL</abbr></li>
