<h3 style="text-align:center; color:skyblue; text-decoration: underline  ; font-weight: bold">csv</h3>

<ol style="color:#0fb3ff ; display:flex ; flex-direction: column ; gap:15px">
<li><abbr title="Reads CSV data from a file as lists">csv.reader(file_object)</abbr></li>
<li><abbr title="Writes CSV data to a file from lists">csv.writer(file_object)</abbr></li>
<li><abbr title="Reads CSV data as dictionaries (column names as keys)">csv.DictReader(file_object)</abbr></li>
<li><abbr title="Writes CSV data from dictionaries">csv.DictWriter(file_object, fieldnames=...)</abbr></li>
<li><abbr title="Defines how CSV fields are separated">delimiter=',' (parameter in reader/writer)</abbr></li>
<li><abbr title="Specifies quoting behavior for fields">quoting=csv.QUOTE_MINIMAL / QUOTE_ALL etc.</abbr></li>
<li><abbr title="Skips header row when writing CSV (in DictWriter)">writeheader()</abbr></li>

</ol>
