<div id="typing">

<h3 style="text-align:center; color:skyblue; text-decoration: underline  ; font-weight: bold">typing</h3>

Used for type hints and annotations, especially important when writing clean, maintainable, and well-documented code.

<ol style="color:#0fb3ff ; display:flex ; flex-direction: column ; gap:15px">
<li><abbr title="Specifies a general type">Any</abbr></li>
<li><abbr title="Used to hint optional (can be None) types">Optional</abbr></li>
<li><abbr title="Defines a list with elements of a specific type">List[type]</abbr></li>
<li><abbr title="Defines a dictionary with specific key and value types">Dict[key_type, value_type]</abbr></li>
<li><abbr title="Defines a tuple with fixed number of element types">Tuple[type1, type2]</abbr></li>
<li><abbr title="Defines a function signature with input and output types">Callable[[arg1_type, arg2_type], return_type]</abbr></li>
<li><abbr title="Defines a sequence type (like list, tuple, etc.)">Sequence[type]</abbr></li>
<li><abbr title="Defines a set with specific element types">Set[type]</abbr></li>
<li><abbr title="Defines a generator function return type">Generator[yield_type, send_type, return_type]</abbr></li>
<li><abbr title="Used when you want to say 'this can be one of these types'">Union[type1, type2]</abbr></li>
<li><abbr title="Used to define generic types (like T in a function)">TypeVar</abbr></li>

</ol>
